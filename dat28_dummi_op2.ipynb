```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np
from statsmodels.stats.outliers_influence import variance_inflation_factor
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import pylab as py

from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
df_GPA=pd.read_csv('GPA_data.csv')
df_GPA.head()
```

```python
le=LabelEncoder()
```

```python
df_GPA[['Gender','Extra_Curricular']]=df_GPA[['Gender','Extra_Curricular']].apply(lambda col: le.fit_transform(col))
df_GPA.head()
```

```python
# since Extra_curicular column has 3 types of catogaries there are additional operations to be done
ohe=OneHotEncoder()
```

```python
arr=ohe.fit_transform(df_GPA['Extra_Curricular'].values.reshape(len(df_GPA['Extra_Curricular'].values),1)).toarray()
```

```python
# should identify the each array column reprecenting for 1st - no ,2nd - societies, 3rd-sport just surfig the data
# as did earlier one column should have required to drop - arr[:,1:]
# this array is represented float values those should be convert to int arr[:,1:].astype(int)
#create a data fram from updated array and named the column respective name 
arr_data=pd.DataFrame(arr[:,1:].astype(int),columns=['societies','sport'])
arr_data.head()
```

```python
# combine with main df
updated_GPA=pd.concat([df_GPA,arr_data],axis=1)
#drop columns
updated_GPA.drop('Extra_Curricular',axis=1,inplace=True)
updated_GPA.head()
```
