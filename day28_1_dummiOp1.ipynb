```python
df_GPA=pd.read_csv('GPA_data.csv')
df_GPA.head()
```

```python
#make dummie df of Gender and Extra_Cur
df_dummies=pd.get_dummies(df_GPA[['Gender','Extra_Curricular']])
df_dummies.head()
```

```python
# drop similer dummies 
df_dummies.drop(['Gender_Female','Extra_Curricular_Sports'],axis=1,inplace=True)
```

```python
# merge df_dummis df to main data frame 
df_GPA=pd.concat([df_GPA,df_dummies],axis=1)
df_GPA.head()
```

```python
#drop catogarical columns 
df_GPA.drop(['Gender','Extra_Curricular'],axis=1,inplace=True)
```
